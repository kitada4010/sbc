```python
number='C14'
epi='restraint'
part='L'

peak_line=-0.08
skip_list=1

imfile_name=epi+'\\'+number+' '+epi+' '+part
exfile_name=epi+'\\'+number+' '+epi+' '+part+' '+'data'
```

```python
im_file='C:\\Users\\ishikawa\\Cross Match\\import file\\'+imfile_name+'\\'
ex_file='C:\\Users\\ishikawa\\Cross Match\\import file\\'+exfile_name+'\\'
```

```python
im_R2=number+' '+epi+' '+part+'2.xlsx'
im_R3=number+' '+epi+' '+part+'3.xlsx'
im_R4=number+' '+epi+' '+part+'4.xlsx'
im_R5=number+' '+epi+' '+part+'5.xlsx' 

ex=number+' '+epi+' '+part+' '+'data.xlsx'
```

```python
%%time
import openpyxl
import pandas as pd
wb_ex=openpyxl.load_workbook(ex_file+ex)
sheet_ex=wb_ex['Sheet1']

wb_im_R2=openpyxl.load_workbook(im_file+im_R2)
for i,p in zip((wb_im_R2.sheetnames),range(len(wb_im_R2.sheetnames)+1)):
    sheet_im_R2=wb_im_R2[i]
    df=pd.read_excel(im_file+im_R2, sheet_name=i, skiprows=skip_list, header=None)
    df=df[1]
    amp=max(df)-min(df)
    dur=len(df)*0.04
    arc_list=[]
    for j in range(len(df)-1):
        arc_piece=((df[j+1]-df[j])**2+0.04**2)**0.5
        arc_list.append(arc_piece)
    arc=sum(arc_list)
    peak_list=[]
    for k in range(3,len(df)-3):
        if df[k]<peak_line and \
           df[k]<df[k-3] and df[k]<df[k-2] and df[k]<df[k-1] and \
           df[k]<df[k+3] and df[k]<df[k+2] and df[k]<df[k+1]:
            peak_piece=1
            peak_list.append(peak_piece)
        else:
            peak_piece=0
            peak_list.append(peak_piece)
    peak=sum(peak_list)
    sheet_ex.cell(row=1, column=1).value='base'
    sheet_ex.cell(row=1, column=2).value='Duration'
    sheet_ex.cell(row=1, column=3).value='Amplitude'
    sheet_ex.cell(row=1, column=4).value='Arc Length'
    sheet_ex.cell(row=1, column=5).value='Peak'
    sheet_ex.cell(row=p+2, column=2).value=dur
    sheet_ex.cell(row=p+2, column=3).value=amp
    sheet_ex.cell(row=p+2, column=4).value=arc
    sheet_ex.cell(row=p+2, column=5).value=peak

wb_ex.save(ex_file+ex)


wb_im_R3=openpyxl.load_workbook(im_file+im_R3)
for i,p in zip((wb_im_R3.sheetnames),range(len(wb_im_R3.sheetnames)+1)):
    sheet_im_R3=wb_im_R3[i]
    df=pd.read_excel(im_file+im_R3, sheet_name=i, skiprows=skip_list, header=None)
    df=df[1]
    amp=max(df)-min(df)
    dur=len(df)*0.04
    arc_list=[]
    for j in range(len(df)-1):
        arc_piece=((df[j+1]-df[j])**2+0.04**2)**0.5
        arc_list.append(arc_piece)
    arc=sum(arc_list)
    peak_list=[]
    for k in range(3,len(df)-3):
        if df[k]<peak_line and \
           df[k]<df[k-3] and df[k]<df[k-2] and df[k]<df[k-1] and \
           df[k]<df[k+3] and df[k]<df[k+2] and df[k]<df[k+1]:
            peak_piece=1
            peak_list.append(peak_piece)
        else:
            peak_piece=0
            peak_list.append(peak_piece)
    peak=sum(peak_list)
    sheet_ex.cell(row=1, column=7).value='20'
    sheet_ex.cell(row=1, column=8).value='Duration'
    sheet_ex.cell(row=1, column=9).value='Amplitude'
    sheet_ex.cell(row=1, column=10).value='Arc Length'
    sheet_ex.cell(row=1, column=11).value='Peak'
    sheet_ex.cell(row=p+2, column=8).value=dur
    sheet_ex.cell(row=p+2, column=9).value=amp
    sheet_ex.cell(row=p+2, column=10).value=arc
    sheet_ex.cell(row=p+2, column=11).value=peak

wb_ex.save(ex_file+ex)


wb_im_R4=openpyxl.load_workbook(im_file+im_R4)
for i,p in zip((wb_im_R4.sheetnames),range(len(wb_im_R4.sheetnames)+1)):
    sheet_im_R4=wb_im_R4[i]
    df=pd.read_excel(im_file+im_R4, sheet_name=i, skiprows=skip_list, header=None)
    df=df[1]
    amp=max(df)-min(df)
    dur=len(df)*0.04
    arc_list=[]
    for j in range(len(df)-1):
        arc_piece=((df[j+1]-df[j])**2+0.04**2)**0.5
        arc_list.append(arc_piece)
    arc=sum(arc_list)
    peak_list=[]
    for k in range(3,len(df)-3):
        if df[k]<peak_line and \
           df[k]<df[k-3] and df[k]<df[k-2] and df[k]<df[k-1] and \
           df[k]<df[k+3] and df[k]<df[k+2] and df[k]<df[k+1]:
            peak_piece=1
            peak_list.append(peak_piece)
        else:
            peak_piece=0
            peak_list.append(peak_piece)
    peak=sum(peak_list)
    sheet_ex.cell(row=1, column=13).value='30'
    sheet_ex.cell(row=1, column=14).value='Duration'
    sheet_ex.cell(row=1, column=15).value='Amplitude'
    sheet_ex.cell(row=1, column=16).value='Arc Length'
    sheet_ex.cell(row=1, column=17).value='Peak'
    sheet_ex.cell(row=p+2, column=14).value=dur
    sheet_ex.cell(row=p+2, column=15).value=amp
    sheet_ex.cell(row=p+2, column=16).value=arc
    sheet_ex.cell(row=p+2, column=17).value=peak

wb_ex.save(ex_file+ex)


wb_im_R5=openpyxl.load_workbook(im_file+im_R5)
for i,p in zip((wb_im_R5.sheetnames),range(len(wb_im_R5.sheetnames)+1)):
    sheet_im_R5=wb_im_R5[i]
    df=pd.read_excel(im_file+im_R5, sheet_name=i, skiprows=skip_list, header=None)
    df=df[1]
    amp=max(df)-min(df)
    dur=len(df)*0.04
    arc_list=[]
    for j in range(len(df)-1):
        arc_piece=((df[j+1]-df[j])**2+0.04**2)**0.5
        arc_list.append(arc_piece)
    arc=sum(arc_list)
    peak_list=[]
    for k in range(3,len(df)-3):
        if df[k]<peak_line and \
           df[k]<df[k-3] and df[k]<df[k-2] and df[k]<df[k-1] and \
           df[k]<df[k+3] and df[k]<df[k+2] and df[k]<df[k+1]:
            peak_piece=1
            peak_list.append(peak_piece)
        else:
            peak_piece=0
            peak_list.append(peak_piece)
    peak=sum(peak_list)
    sheet_ex.cell(row=1, column=19).value='40'
    sheet_ex.cell(row=1, column=20).value='Duration'
    sheet_ex.cell(row=1, column=21).value='Amplitude'
    sheet_ex.cell(row=1, column=22).value='Arc Length'
    sheet_ex.cell(row=1, column=23).value='Peak'
    sheet_ex.cell(row=p+2, column=20).value=dur
    sheet_ex.cell(row=p+2, column=21).value=amp
    sheet_ex.cell(row=p+2, column=22).value=arc
    sheet_ex.cell(row=p+2, column=23).value=peak

wb_ex.save(ex_file+ex)
```
