```python
#エクセルファイルの変換を目指して
import pandas as pd
file = pd.ExcelFile('B39 restraint R2.xls')
#all_df = pd.DataFrame()


sheet_df = file.parse(file.sheet_names, header=None)
all_df = pd.DataFrame()

sheet_names = file.sheet_names
for i, name in enumerate(sheet_names):
    sheet_df[i] = file.parse(name)
    #print(sheet_df[i]['INFORMATION'][20:])
    all_df = pd.concat([all_df, sheet_df[i]['INFORMATION'][21:]], axis=1)

file.close()

print(all_df)

import xlrd
file = xlrd.open_workbook('B39 restraint R2.xls')
print(file.nsheets)

all_df.to_csv("otamesi.csv", header=False, index=False)
```

```python
#自己相関を求めるために
import pandas as pd
import numpy as np
deta = pd.read_csv('otamesi.csv')
#print(deta)

#print(deta[:,0])
sig1 = deta.values[:,0]
sig2 = deta.values[:,1]
sig1_float = sig1.astype(np.float32)
print(sig1_float)


#sig1 = sig1[~np.isnan(sig1).any(axis=1)]
print(sig1[~np.isnan(sig1)])
print(sig1)
  
#np_data[0] = np_data[0] - np_data[0].mean()
#np_data[1] = np_data[1] - np_data[1].mean()
sig1 = sig1 - sig1.mean()
sig2 = sig2 - sig2.mean()

corr = np.correlate(sig1, sig2, "full")
estimated_delay = corr.argmax() - (len(sig1) - 1)
print("estimated delay is " + str(estimated_delay))
```
